## Create Train test data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("price" , axis =1)
Y = df['price']

#SPlot in ti 80:20

X_train, x_test, Y_train , y_test = train_test_split(X,Y,test_size = 0.2 , random_state=0)
```

## Train the model

```python
# Linear Regeression
from sklearn.linear_model import LinearRegression

liner_model = LinearRegression()
liner_model.fit(X_train, Y_train)
```

```python
liner_model.score(X_train, Y_train)
```

```python
predeictors = X_train.columns
coef = pd.Series(liner_model.coef_ , predeictors).sort_values()
```

```python
print(coef)
```

## Test Linear


```python
y_predict = liner_model.predict(x_test)
```

```python
r_sq = liner_model.score(x_test, y_test)
r_sq
```

```python
y_test
```

```python
y_predict
```

```python
from sklearn.metrics import mean_squared_error

lm_mse = mean_squared_error(y_predict , y_test)
lm_mse
```

```python
plt.plot(y_predict, label = "Predicted")
plt.plot(y_test.values , label = "Actual")
plt.legend()
plt.show()
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha = 0.5 , normalize = True)
lasso_model.fit(X_train, Y_train)
```

```python
# R SQ
lasso_model.score(X_train, Y_train)
```

```python
predeictors = X_train.columns
coef = pd.Series(lasso_model.coef_ , predeictors).sort_values()
print(coef)
```

## Test Lasso

```python
y_predict = lasso_model.predict(x_test)
```

```python
r_sq = lasso_model.score(x_test, y_test)
r_sq
```

```python
plt.plot(y_predict, label = "Predicted")
plt.plot(y_test.values , label = "Actual")
plt.legend()
plt.show()
```

```python
lm_mse = mean_squared_error(y_predict , y_test)
math.sqrt(lm_mse)
```

##  Ridge Regression
