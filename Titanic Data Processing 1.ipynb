##  Replace with Median Age of title

```python
df.Name
```

```python
def GetTitle(name):
    title_group = {        
       'mr':'Mr','mrs':'Mrs', 'miss':'Miss', 'master':'Master', 'don':'Sir', 'rev':'Sir', 'dr':'Officer', 'mme':'Mrs', 'ms':'Mrs',
       'major':'Officer', 'lady':'Lady', 'sir':'Sir', 'mlle':'Miss', 'col':'Officer', 'capt':'Officer', 'the countess':'Lady',
       'jonkheer':'Sir', 'dona':'Lady'
    }
    
    nameWithTitle = name.split(",")[1]
    title = nameWithTitle.split(".")[0]
    title = title.strip().lower()
    return title_group[title]
```

```python
GetTitle("Minahan, Mrs. William Edward (Lillian E Thorpe)")
```

```python
df.Name.map(lambda x : GetTitle(x))
```

```python
df.Name.map(lambda x : GetTitle(x)).unique()
```

```python
df["Title"] = df.Name.map(lambda x : GetTitle(x))
```
